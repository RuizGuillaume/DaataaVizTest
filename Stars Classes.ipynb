```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sample_data/stars_classes.csv', sep=",")
df.head()
```

```python
x_classes = ['O','B','A','F','G','K','M']

plt.rcParams['axes.facecolor'] = '202020'

df.plot(x="Temperature (K)", y="Luminosity(L/Lo)", kind="scatter",
        c="Temperature (K)", colormap="RdYlBu", s=4,
        title="Classes stellaires")

plt.gca().invert_xaxis()
plt.yscale("log")

plt.show()
```
